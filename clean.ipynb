```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib as mpl
```

```python
df = pd.read_csv("dataset\credit_risk_dataset.csv")
df
```

```python
df["person_age"].max()
```

```python
df.describe()
```

```python
df.isna().sum()
```

```python
# Check correlation and visualize BEFORE cleaning
correlation_before = df['person_age'].corr(df['person_emp_length'])
print(f"Correlation between Age and Employment Length BEFORE cleaning: {correlation_before:.4f}")

plt.figure(figsize=(10, 6))
sns.scatterplot(x='person_age', y='person_emp_length', data=df, alpha=0.5, color='#B74A4A')
plt.title("Correlation between Age and Employment Length (Before Cleaning)")
plt.xlabel("Person Age")
plt.ylabel("Employment Length (Years)")
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
df = df.dropna()
```

```python
df.describe()
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x=df['person_age'], color="#B74A4A") 
plt.title("Person Age Distribution Box Plot")
plt.xlabel("Age")
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
print(df['person_age'].max())

clients_over_80 = df[df['person_age'] > 80]
print(f"older than 80  : {len(clients_over_80)}")

df = df[df['person_age'] <= 80]

print(df.shape)
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x=df['person_age'], color="#5DA5DA")
plt.title("Person Age Distribution Box Plot")
plt.xlabel("Age")
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
x_axis = df['person_emp_length'].value_counts().index
y_axis = df['person_emp_length'].value_counts().values
fig = plt.figure(figsize=(10,5))
plt.bar(x_axis,y_axis,color="#CFA027")
plt.grid(True, linestyle='--', alpha=0.6)
```

```python
df = df.drop(df[df['person_emp_length']>40].index,axis=0)
```

```python

correlation = df['person_age'].corr(df['person_emp_length'])
print(f"Correlation between Age and Employment Length: {correlation:.4f}")


plt.figure(figsize=(10, 6))
sns.scatterplot(x='person_age', y='person_emp_length', data=df, alpha=0.5, color='#4CAF50')
plt.title("Correlation between Age and Employment Length")
plt.xlabel("Person Age")
plt.ylabel("Employment Length (Years)")
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
print(df['cb_person_default_on_file'].value_counts())

df['cb_person_default_on_file'] = df['cb_person_default_on_file'].apply(lambda x: 1 if x == 'Y' else 0)

print("\nAfter mapping:")
print(df['cb_person_default_on_file'].value_counts())
```
